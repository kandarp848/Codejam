```python
import numpy as np
import sys
from itertools import combinations

def complement(bits): return [
    '_' if bu is '_' else
    '1' if bu=='0' else '0'
    for bu in bits
]

class Barray():
    def __init__(self, Bi):
        self.bits = ['_'] * Bi
        self.unkowns = list(range(Bi))
        self.read(10)
        self.run()

    def read(self, ni):
        for _ in range(ni):
            i = self.unkowns.pop()
            self.bits[i] = self.read_bit(i+1)
            self.unkowns = self.unkowns[::-1]
        self.update_state()

    def read_bit(self, ilo):
        print(ilo, flush=True)
        return input()

    def update_state(self):
        self.bits_c  = complement(self.bits)
        self.bits_cr = self.bits_c[::-1]
        self.bits_r  = complement(self.bits_cr)
        self.states = [self.bits, self.bits_c, self.bits_cr, self.bits_r]

    def get_test(self):
        candids = list(set(range(Bi)) - set(self.unkowns))
        max_states = len(set(map(tuple, self.states)))
        for idx in combinations(candids, 2):
            num_states = len(set(
                tuple(np.take(state, idx))
                for state in self.states
            ))
            if num_states == max_states: return idx

    def collap(self):
        test_idx = self.get_test()
        test = [self.read_bit(i+1) for i in test_idx]

        self.bits = next(state
            for state in self.states
            if test == list(np.take(state, test_idx))
        )

    def run(self):
        while True:
            self.collap()
            try: self.read(8)
            except IndexError: break


t, Bi = map(int, input().split())
print('B:', Bi, file=sys.stderr)
for _ in range(t):
    array = Barray(Bi)
    print('guess:', ''.join(array.bits), file=sys.stderr)
    print(''.join(array.bits), flush=True)
    if input() == 'N': sys.exit()
```
